In this analysis we will analyze which news was seen and read the most by the number of hits it got

```python
import os
import nltk
import string
from nltk.corpus import stopwords
import operator
from glob import glob
import json
from operator import itemgetter
```

```python
from os.path import expanduser
home = expanduser("~")
root = home+"//midterm//data//Article_Search//"
```

```python
col = []
for filename in glob(root+"*.json"):
    with open(filename) as data_file:
        data = json.load(data_file)
    hits = data["response"]["meta"]["hits"]
    for at in data["response"]["docs"]:
        nd = at["headline"]["main"]
    col.append((nd,hits))
    
sorted_data = sorted(col, key=lambda x : x[1],reverse=True)
for x in sorted_data:
    print(x[:25])
```

```python
import csv
csvFile = open('MostHits.csv', 'w')
writer = csv.writer(csvFile)
writer.writerow(['Article', 'Rank', 'Hits']) #header
for i in range(len(sorted_data)):
    addRank = lambda sorted_data,i: [sorted_data[i][0], i+1, sorted_data[i][1]] #lambda 
    writer.writerow(addRank(sorted_data,i))
csvFile.close()
```

# Conclusion:
Based on the number of hits an article got we can conclude that 'In 3 New Travel Series, the Comic, the Solemn and the Vroom!' and 'Gordie Howe, Mr. Hockey for the Red Wings, Dies at 88' were the most read and seen sports article
