# Question 1 Analysis 1
In this Analysis we will loop through all the json files having articles ranging from 2000 to 2017 and will take out the count on the type of news that were published in New York times

```python
import os
import operator #used for sorting
import string
from glob import glob
import datetime
from datetime import timezone
import json
from collections import Counter #to count frequency
```

```python
from os.path import expanduser
home = expanduser("~")
root = home+"\\midterm\\data\\Archive"
print("Root :",root)
```

```python
Y = list(range(2000,2017)) #this will iterate from year 2000 to 2017
for y in Y:
    for filename in glob(root+"\\"+str(y)+"\\*.json"):
        with open(filename) as data_file:
            data = json.load(data_file)
        news_desk = {}
        col = []
        for at in data["response"]["docs"]:
            nd = at["news_desk"]    #will take the type of news in each article
            if(nd != None):         #if news type is none it will be ignored
                col.append(nd)
import nltk
frequency = nltk.FreqDist(col)
print(frequency)
```

```python
frequency.plot(30,cumulative=False)
```

# Conclusion:
From the above plot it is visible that business types of news was published the most from 2000 to 2017
