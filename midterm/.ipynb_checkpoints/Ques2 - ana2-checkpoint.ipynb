```python
import os
import nltk
import string
from nltk.corpus import stopwords
import operator
from glob import glob
import json
```

```python
hdlst1 = list()
hdlst2 = list()
hdlst3 = list()
hdlst4 = list()
wrdlst1 = list()
wrdlst2 = list()
wrdlst3 = list()
wrdlst4 = list()
word1 = "federer"
wrdlst1.append(word1)
word2 = "djokovic"
word3="nadal"
wrdlst3.append(word3)
word4="murray"
wrdlst4.append(word4)
wrdlst2.append(word2)
[dataProcessing(year,month,wrdlst1,wrdlst2,wrdlst3,wrdlst4) for year in reversed(range(2014,2016)) for month in range(1,13)]
print("There were %s number of headlines related to %s,%s for %s,%s for %s and %s for %s"%(len(hdlst1),word1,len(hdlst2),word2,len(hdlst3),word3,len(hdlst4),word4))
```

```python
from os.path import expanduser
home = expanduser("~")
root = home+"//midterm//data//archive//"
```

```python
def dataProcessing(year,month,list1,list2,list3,list4):
    path = root+str(year)+"//"+str(month)+".json"
    with open(path) as data_file:
        data = json.load(data_file)
    [headlineAnalysis(at,list1,list2,list3,list4) for at in data["response"]["docs"]]
```

```python
def headlineAnalysis(ar,list1,list2,list3,list4):
    try:
        headline = ar["headline"]["main"]
        headline = headline.lower()
        headline = removePunctuations(headline)
        headline = [word for word in headline.split(" ") if word not in stopwords.words('english')]
        l1 = len([i for i in headline if i in list1])
        if(l1 > 0):
            hdlst1.append(ar["headline"]["main"])
        l2 = len([i for i in headline if i in list2])
        if(l2 > 0):
            hdlst2.append(ar["headline"]["main"])
        l3 = len([i for i in headline if i in list3])
        if(l3 > 0):
            hdlst3.append(ar["headline"]["main"])
        l4 = len([i for i in headline if i in list4])
        if(l1 > 0):
            hdlst4.append(ar["headline"]["main"])
    except:
        pass
```

```python
def removePunctuations(word):
    rmvp = str.maketrans('','',string.punctuation)
    return word.translate(rmvp)
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
 
objects = ('Roger Federer', 'Novak Djokovic','Rafael Nadal','Andy Murray')
y_pos = np.arange(len(objects))
performance = [len(hdlst1),len(hdlst2),len(hdlst3),len(hdlst4)]
 
plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Count')
plt.title('Headlines made by Federer, Djokovic, Nadal and Murray in 2014-2015')
 
plt.show()
```
