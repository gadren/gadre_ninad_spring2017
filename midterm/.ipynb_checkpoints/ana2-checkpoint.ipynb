# Question 1 Analysis 2
Kenneth Lay, Jeffrey Skilling, John forney and David Delainey were the main convicts behind the Enron Scandal. In this analysis we would go through all the folders of all 4 of them and find out top 10 people from whom they got emails and top 10 people to whom they emailed

```python
import os
import email #used for email parsing
import operator #used for sorting data
import json
from glob import glob
```

```python
root = "C:\\Users\\ninad\\midterm\\data\\enron\\maildir"
print("Root: ",root)
```

```python
#This will loop  get only the email addresses from the whole email
def getComData(user): #user will be the person whose folders are to be scanned
    comData = {"user":user, "coms":{}}
    for subdir, dirs, files in os.walk(user):
        for file in files:
            if(file[0]!="."):#prevents hidden files from running
                try:
                    emf = open(os.path.join(subdir, file),"r", errors='ignore')
                    em = email.message_from_string(emf.read())
                except:
                    print("EROR",os.path.join(subdir, file))
                for to in str(em['to']).split(","):
                    if( to.strip() in comData["coms"]):
                        comData["coms"][to.strip()] = comData["coms"][to.strip()] + 1
                    else:
                         comData["coms"][to.strip()] = 1
    return comData
```

```python
#This method will sort the data based on count the email address was mentioned
def printTop(comData,x):
    sorted_data = sorted(comData["coms"].items(), key=operator.itemgetter(1), reverse=True)
    print("User: ",comData['user'].split("/"))
    for com in sorted_data[:x]:
        print("\t",com)
```

```python
for x in ["lay-k","skilling-j","delainey-d","forney-j"]:  #I have mentioned only these 4 names as that is the data I want to have a look at
    for dirct in glob(root+"\\"+str(x)+"\\*\\"):
        printTop(getComData(dirct),10)
```
