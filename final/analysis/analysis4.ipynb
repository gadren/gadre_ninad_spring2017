```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_by_state_1997 = df_1997.groupby('DEST_STATE_NM')['YEAR'].count()
df_by_state_1987 = df_1987.groupby('DEST_STATE_NM')['YEAR'].count()
df_by_state_2016 = df_2016.groupby('DEST_STATE_NM')['YEAR'].count()



print (len(df_by_state_1997))
print (len(df_by_state_1987))
print (len(df_by_state_2016))

print (df_by_state_1987)
```

```python
plt.figure(figsize=(15,6))

df_by_state_1997.sort_index().plot(c='b',label='1997',use_index='True',xticks=[i for i in range(52)])
df_by_state_1987.sort_index().plot(c='r',label="1987",sharex='True')
df_by_state_2016.sort_index().plot(c='g',label="2016",sharex='True')

plt.legend()
plt.xticks(rotation='vertical')

print (len(df_by_state_1997.index))
```

```python
df_by_state_2016 = df_2016[df_2016.DEST_STATE_NM=='California'].groupby('DEST_CITY_NAME')['YEAR'].count()
df_by_state_2016
```

```python
df_by_state_1997 = df_1997[df_1997.DEST_STATE_NM=='California'].groupby('DEST_CITY_NAME')['YEAR'].count()
df_by_state_1997
```

```python
df_by_state_1987 = df_1987[df_1987.DEST_STATE_NM=='California'].groupby('DEST_CITY_NAME')['YEAR'].count()
df_by_state_1987
```

```python
plt.figure(figsize=(15,5))
temp = pd.concat([df_by_state_1997,df_by_state_1987,df_by_state_2016],axis=1)
temp.columns = ['1997','1987','2016']
temp.head()
temp.plot(kind='bar',figsize=(15,5))
plt.xticks(rotation='vertical')
plt.ylabel('total number of flights')
plt.xticks(range(len(temp)),temp.index)
```
