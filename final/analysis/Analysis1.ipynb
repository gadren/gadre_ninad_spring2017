```python
import pandas as pd
import numpy as np
```

```python
import os
path = os.getcwd()+'\\data\\May2016_Flights.csv'
may = pd.read_csv(path)
may.head()
```

```python
path1 = os.getcwd()+'\\data\\Aug2016_flights.csv'
aug = pd.read_csv(path1)
aug.head()
```

```python
path2 = os.getcwd()+'\\data\\Dec2016_Flights.csv'
dec= pd.read_csv(path2)
dec.head()
```

```python
may = may.drop(may.columns[7],axis=1)
aug = aug.drop(aug.columns[7],axis=1)
dec = dec.drop(dec.columns[7],axis=1)
```

```python
MayCancel = may.groupby(by=['ORIGIN'])['CANCELLED'].sum()
MayCancel.sort_values(inplace=True, ascending=False)
MayCancel[:5]
```

```python
AugCancel = aug.groupby(by=['ORIGIN'])['CANCELLED'].sum()
AugCancel.sort_values(inplace=True, ascending=False)
AugCancel[:5]
```

```python
DecCancel = dec.groupby(by=['ORIGIN'])['CANCELLED'].sum()
DecCancel.sort_values(inplace=True, ascending=False)
DecCancel[:5]
```

```python
MayNumFlights = may['ORIGIN'].value_counts()
MayNumFlights[:5]
```

```python
AugNumFlights = aug['ORIGIN'].value_counts()
AugNumFlights[:5]
```

```python
DecNumFlights = dec['ORIGIN'].value_counts()
DecNumFlights[:5]
```

```python
MayCancelf = (MayCancel/MayNumFlights)*100
MayCancelf.sort_values(inplace=True, ascending=False)
MayCancelf[:5]
```

```python
AugCancelf = (AugCancel/AugNumFlights)*100
AugCancelf.sort_values(inplace=True, ascending=False)
AugCancelf[:5]
```

```python
DecCancelf = (DecCancel/DecNumFlights)*100
DecCancelf.sort_values(inplace=True, ascending=False)
DecCancelf[:5]
```

```python
MayDelay = may.groupby(by=['ORIGIN'])['DEP_DEL15'].sum()
MayDelay.sort_values(inplace=True, ascending=False)
MayDelay[:5]
```

```python
MayDelayf = MayDelay/MayNumFlights
MayDelayf.sort_values(inplace=True, ascending=False)
MayDelayf[:5]
```

```python
AugDelay = aug.groupby(by=['ORIGIN'])['DEP_DEL15'].sum()
AugDelay.sort_values(inplace=True, ascending=False)
AugDelay[:5]
```

```python
AugDelayf = AugDelay/AugNumFlights
AugDelayf.sort_values(inplace=True, ascending=False)
AugDelayf[:5]
```

```python
DecDelay = dec.groupby(by=['ORIGIN'])['DEP_DEL15'].sum()
DecDelay.sort_values(inplace=True, ascending=False)
DecDelay[:5]
```

```python
DecDelayf = DecDelay/DecNumFlights
DecDelayf.sort_values(inplace=True, ascending=False)
DecDelayf[:5]
```

```python
path3 = os.getcwd()+'\\data\\airports new2.csv'
airport = pd.read_csv(path3)
airport.head()
```

```python
MayAirpts = pd.concat([MayCancel, MayCancelf, MayDelay, MayDelayf], axis=1)
MayAirpts.head()
```

```python
AugAirpts = pd.concat([AugCancel, AugCancelf, AugDelay, AugDelayf], axis=1)
AugAirpts.head()
```

```python
DecAirpts = pd.concat([DecCancel, DecCancelf, DecDelay, DecDelayf], axis=1)
DecAirpts.head()
```

```python
MayAirpts.columns = ['numCancel', 'fracCancel','numDelay','fracDelay']
MayAirpts['iata'] = MayAirpts.index
MayAirpts.head()
```

```python
AugAirpts.columns = ['numCancel', 'fracCancel','numDelay','fracDelay']
AugAirpts['iata'] = AugAirpts.index
AugAirpts.head()
```

```python
DecAirpts.columns = ['numCancel', 'fracCancel','numDelay','fracDelay']
DecAirpts['iata'] = DecAirpts.index
DecAirpts.head()
```

```python
MayAirLoc = pd.merge(airport, MayAirpts, on='iata', how='inner')
MayAirLoc.head()
```

```python
AugAirLoc = pd.merge(airport, AugAirpts, on='iata', how='inner')
AugAirLoc.head()
```

```python
DecAirLoc = pd.merge(airport, DecAirpts, on='iata', how='inner')
DecAirLoc.head()
```

```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
from matplotlib import rcParams
plt.rc('font', family='serif')
plt.rc('font', serif='Computer Modern Roman')
plt.rc('text', usetex='false')
rcParams['axes.labelsize'] = 15
rcParams['xtick.labelsize'] = 15
rcParams['ytick.labelsize'] = 15
rcParams['legend.fontsize'] = 15
rcParams['axes.titlesize'] = 15
```

```python
plt.figure(figsize=(12,8))

plt.subplot(221)
plt.title('Number of Cancelled Flights')
MayAirLoc.numCancel.hist(label='May')
AugAirLoc.numCancel.hist(alpha=0.4, label='August')
DecAirLoc.numCancel.hist(alpha=0.4, label='December')
plt.legend()

plt.subplot(222)
plt.title('Number of Delayed Flights')
MayAirLoc.numDelay.hist()
AugAirLoc.numDelay.hist(alpha=0.4)
DecAirLoc.numDelay.hist(alpha=0.4)

plt.subplot(223)
plt.title('Fraction of Cancelled Flights')
MayAirLoc.fracCancel.hist()
AugAirLoc.fracCancel.hist(alpha=0.6)
DecAirLoc.fracCancel.hist(alpha=0.6)

plt.subplot(224)
plt.title('Fraction of Delayed Flights')
MayAirLoc.fracDelay.hist()
AugAirLoc.fracDelay.hist(alpha=0.6)
DecAirLoc.fracDelay.hist(alpha=0.6)

plt.tight_layout()
```

```python
from mpl_toolkits.basemap import Basemap
colorScale = plt.cm.get_cmap('jet')

plt.figure(figsize=(24,16))

# Map of U.S. lower 48 adapted from fillstates.py
# available here: https://github.com/matplotlib/basemap/blob/master/examples/fillstates.py

plt.subplot(221)
# Lambert Conformal map of lower 48 states.
m = Basemap(llcrnrlon=-119,llcrnrlat=22,urcrnrlon=-64,urcrnrlat=49,
            projection='lcc',lat_1=33,lat_2=45,lon_0=-95,
            resolution='l',area_thresh=100000.0)

m.drawcoastlines()
m.drawstates()
m.drawcountries()

m.fillcontinents(color = 'white', zorder=0)
m.drawmapboundary(fill_color='steelblue')

x,y = m(MayAirLoc['long'].values, MayAirLoc['lat'].values)

m.scatter(x, y, s=MayAirLoc['fracCancel']*1500., c=MayAirLoc['fracDelay']*100., cmap=colorScale, alpha=0.75)
plt.title('Fraction of Cancelled and Delayed Flights in May')
cbar = plt.colorbar()
cbar.set_label('Percentage of Flights Delayed')

plt.subplot(222)
m = Basemap(llcrnrlon=-119,llcrnrlat=22,urcrnrlon=-64,urcrnrlat=49,
            projection='lcc',lat_1=33,lat_2=45,lon_0=-95,
            resolution='l',area_thresh=100000.0)

m.drawcoastlines()
m.drawstates()
m.drawcountries()

m.fillcontinents(color = 'white', zorder=0)
m.drawmapboundary(fill_color='steelblue')

x,y = m(AugAirLoc['long'].values, AugAirLoc['lat'].values)

m.scatter(x, y, s=AugAirLoc['fracCancel']*1500., c=AugAirLoc['fracDelay']*100., cmap=colorScale, alpha=0.75)
plt.title('Fraction of Cancelled and Delayed Flights in August')
cbar = plt.colorbar()
cbar.set_label('Percentage of Flights Delayed')

plt.subplot(223)
m = Basemap(llcrnrlon=-119,llcrnrlat=22,urcrnrlon=-64,urcrnrlat=49,
            projection='lcc',lat_1=33,lat_2=45,lon_0=-95,
            resolution='l',area_thresh=100000.0)

m.drawcoastlines()
m.drawstates()
m.drawcountries()

m.fillcontinents(color = 'white', zorder=0)
m.drawmapboundary(fill_color='steelblue')

x,y = m(DecAirLoc['long'].values, DecAirLoc['lat'].values)

m.scatter(x, y, s=DecAirLoc['fracCancel']*1500., c=DecAirLoc['fracDelay']*100., cmap=colorScale, alpha=0.75)
plt.title('Fraction of Cancelled and Delayed Flights in December')
cbar = plt.colorbar()
cbar.set_label('Percentage of Flights Delayed')

plt.tight_layout()
plt.show()
```
